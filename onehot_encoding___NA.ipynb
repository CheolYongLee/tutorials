```python
import pandas as pd
```

```python
iris_path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris2.csv'
iris = pd.read_csv(iris_path)
```

```python
iris.head()
```

```python
# 원핫 인코딩
encoding = pd.get_dummies(iris)
encoding.head()
```

```python
print(iris.dtypes)
```

```python
# 품종 타입을 범주형으로 바꾸기
iris['품종'] = iris['품종'].astype('category')
print(iris.dtypes)
```

```python
# 원핫 인코딩
encoding = pd.get_dummies(iris)
encoding.head()
```

```python
# NA 값 체크
iris.isna().sum()
```

```python
# 임의로 넣은 NA 값 확인
iris.tail()
```

```python
# NA 값에 꽃잎폭 평균 값을 넘어주는 방법
mean = iris['꽃잎폭'].mean()
iris['꽃잎폭'] = iris['꽃잎폭'].fillna(mean)
iris.tail()
```
