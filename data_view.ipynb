# **데이터 불러오기**

```python
import pandas as pd
```

## csv 읽기


```python
# 파일들로부터 데이터 읽어오기
lemonade_path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv'
lemonade = pd.read_csv(lemonade_path)

boston_house_path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
boston = pd.read_csv(boston_house_path)

iris_path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
iris = pd.read_csv(iris_path)
```

## 데이터 모양 확인


```python
# 데이터 모양(row, column)으로 확인하기
print(lemonade.shape)
print(boston.shape)
print(iris.shape)
```

## 칼럼 이름 츨력

```python
# 칼럼 이름으로 출력
print(lemonade.columns)
print(boston.columns)
print(iris.columns)
```

## 독립 변수와 종속 변수 설정

```python
lemonade_inde = lemonade[['온도']]
lemonade_de = lemonade[['판매량']]
print(lemonade_inde.shape, lemonade_de.shape)

boston_inde = boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
boston_de = boston[['medv']]
print(boston_inde.shape, boston_de.shape)

iris_inde = iris[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
iris_de = iris[['품종']]
print(iris_inde.shape, iris_de.shape)
```
