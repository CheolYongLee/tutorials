# 보스턴 집 값 예측

  *   보스턴 집 값 예측: https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv

```python
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```python
# 데이터 준비
boston_path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
boston = pd.read_csv('https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv')
```

```python
boston.head()
```

```python
# 독립변수와 종속변수 지정
boston_inde = boston[['crim',	'zn',	'indus',	'chas',	'nox',	'rm',	'age',	'dis',	'rad',	'tax',	'ptratio',	'b',	'lstat']]
boston_de = boston[['medv']]
```

```python
print(boston_inde.shape, boston_de.shape)
```

```python
# 모델 생성
X = tf.keras.layers.Input(shape=[13])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```python
# 모델 학습
model.fit(boston_inde, boston_de, epochs=20, verbose=2)
```

```python
# 모델 이용
model.predict(boston_inde[:])
```
