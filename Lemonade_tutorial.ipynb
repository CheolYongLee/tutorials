# 레모네이드 판매 예측

  *   github csv url: https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv




```python
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```python
# 데이터 전처리
lemonade_path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv'
lemonade = pd.read_csv(lemonade_path)
```

```python
lemonade.head()
```

```python
# 독립변수, 종속변수
lemonade_inde = lemonade[['온도']]
lemonade_de = lemonade[['판매량']]
```

```python
print(lemonade_inde.shape, lemonade_de.shape)
```

```python
# 모델 만들기
X = tf.keras.layers.Input(shape=[1])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```python
# 모델을 학습하기
model.fit(lemonade_inde, lemonade_de, epochs=10, verbose=2)
```

```python
# 모델을 이용
model.predict(lemonade_inde)
```

```python
# 모델을 통한 예측
model.predict([[15]])
```
